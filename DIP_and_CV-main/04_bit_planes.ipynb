**Bit planes** of an image refer to the binary representation of its pixel values, broken down into individual bits. For an 8-bit grayscale image, each pixel value is represented by 8 bits (from the most significant bit (MSB) to the least significant bit (LSB)). The bit planes are layers or "slices" formed by isolating each bit position across all pixels.

- **Higher bit planes** (e.g., MSB) carry more visually significant information.
- **Lower bit planes** (e.g., LSB) mostly capture fine details or noise.

This concept is widely used in image processing tasks like compression, watermarking, and steganography.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the image in grayscale

image = cv2.imread('apple.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
#Get dimensions of image
h,w = image.shape
```

```python
#array to store bit planes
bit_planes = np.zeros((8,h,w), dtype=np.uint8)
```

```python
#calculating bit planes
for i in range(8):
  bit_planes[i] = (image>>i)&1
```

```python
#display bit planes
plt.figure(figsize = (12,6))
for i in range(8):
  plt.subplot(2,4,i+1)
  plt.imshow(bit_planes[i],cmap = "gray")
  plt.title(f'Bit Plane {i+1}')
  plt.axis('off')

plt.tight_layout()
plt.show()
```
