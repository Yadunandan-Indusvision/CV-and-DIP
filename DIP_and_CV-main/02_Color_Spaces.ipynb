###What are Colour Spaces?
Color spaces are mathematical models that define how colors are represented in digital form. They define how colors are represented using numerical values, allowing devices (like cameras, monitors, and printers) to reproduce colors accurately.


```python
import cv2
import matplotlib.pyplot as plt

imgPath = "balls.jpg"
image = cv2.imread(imgPath)
```

### 1. **RGB (Red, Green, Blue)**:
   - This is the most widely used color space for digital screens.
   - Colors are created by combining red, green, and blue light in varying intensities.
   - Example: `(255, 0, 0)` represents pure red.
   - operates on an additive model, wherein distinct colors are produced through linear combinations of the red, green, and blue channels.

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(rgb)

plt.figure(figsize=(9,2))

plt.subplot(141)
plt.title('Original')
plt.imshow(rgb)
plt.axis("off")

plt.subplot(142)
plt.title('Red channel')
plt.imshow(r)
plt.axis("off")

plt.subplot(143)
plt.title('Blue channel')
plt.imshow(b)
plt.axis("off")

plt.subplot(144)
plt.title('Green channel')
plt.imshow(g)
plt.axis("off")

plt.savefig('RGB.png')
```

#### Gray color space

For grayscale images, the intensity of a pixel is represented by a single channel, often referred to as the 'gray channel.' This channel stores the luminance information of the image, with higher intensity values indicating lighter shades and lower intensity values representing darker shades.

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)

# Display the original and matched images
plt.figure(figsize=(8, 2))

plt.subplot(1, 2, 1)
plt.imshow(rgb, cmap='gray')
plt.title('Original image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap='gray')
plt.title('Gray scale image')
plt.axis('off')

plt.savefig('GrayScale.png')
```

### HSV color space

HSV color spaces make use of only one channel to describe color i.e. Hue, making it intuitive for specifying colors.

It Designed to be more intuitive for human perception.

**Hue**: The attribute of a color that distinguishes it from other colors. It is often described in terms of degrees around a color wheel and represents the dominant wavelength of light. Hue defines the type of color, such as red, green, blue, etc.

**Saturation**: The intensity or purity of a color. It refers to the amount of gray in proportion to the hue, with lower saturation indicating a more washed-out or pastel-like color, and higher saturation indicating a more vivid and intense color.

**Value(or Brightness)** : The brightness of a color. It represents the relative lightness or darkness of a color, ranging from black (0% brightness) to the maximum brightness of the color (100%). Value determines how light or dark a color appears, regardless of its hue and saturation.



```python
hsv = cv2.cvtColor(rgb, cv2.COLOR_BGR2HSV)

# Splitting the image into different channels
h,s,v = cv2.split(hsv)

# Setting the grid size
plt.figure(figsize=(9,2))

# Plotting the original image
plt.subplot(141)
plt.title('HSV')
plt.imshow(hsv)
plt.axis("off")

# Plotting the channels
plt.subplot(142)
plt.title('Hue channel')
plt.imshow(h)
plt.axis("off")

plt.subplot(143)
plt.title('Saturation channel')
plt.imshow(s)
plt.axis("off")

plt.subplot(144)
plt.title('Value channel')
plt.imshow(v)
plt.axis("off")

plt.savefig('HSV.png')
```

#### YCrCb color space
**Y (Luma)**: The Y channel represents the luminance or brightness of the image. This component is obtained from the RGB color space after gamma correction.

**Cr Channel**: Represents the difference between the R color channel of RGB color space and the luminance component.

**Cb Channel**: Represents the difference between the B color channel of RGB color space and the luminance component.

 - Separation of luminance and chrominance into different color channels is done to optimize color storage.

 - Used in video compression and broadcasting.

```python
yCrCb = cv2.cvtColor(rgb, cv2.COLOR_BGR2YCrCb)

# Splitting the image into different channels
y, cr, cb = cv2.split(yCrCb)

# Setting the grid size
plt.figure(figsize=(9,2))

# Plotting the original image
plt.subplot(141)
plt.title('YCrCb')
plt.imshow(yCrCb)
plt.axis("off")

# Plotting the channels
plt.subplot(142)
plt.title('Y Channel')
plt.imshow(y)
plt.axis("off")

plt.subplot(143)
plt.title('Cr Channel')
plt.imshow(cr)
plt.axis("off")

plt.subplot(144)
plt.title('Cb Channel')
plt.imshow(cb)
plt.axis("off")

plt.savefig('YCrCb.png')
```
