```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("parrot.jpg")
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image, cmap= "gray")
plt.axis("off")
plt.show()
```

```python
# Display each colour channel

redChannel = image.copy()
greenChannel = image.copy()
blueChannel = image.copy()

redChannel[:,:, 1] = 0
redChannel[:,:, 2] = 0

greenChannel[:,:, 0] = 0
greenChannel[:,:, 2] = 0

blueChannel[:,:, 0] = 0
blueChannel[:,:, 1] = 0

plt.figure(figsize=(7,2))

plt.subplot(131)
plt.title('Red channel')
plt.imshow(redChannel)
plt.axis("off")

plt.subplot(132)
plt.title("Green Channel")
plt.imshow(greenChannel)
plt.axis("off")

plt.subplot(133)
plt.title('Blue channel')
plt.imshow(blueChannel)
plt.axis("off")

plt.savefig('ChannelRepresentationColor.png')
```

```python
# Each color channel in a gray scale view

redChannel = cv2.cvtColor(redChannel, cv2.COLOR_BGR2GRAY)
greenChannel = cv2.cvtColor(greenChannel, cv2.COLOR_BGR2GRAY)
blueChannel = cv2.cvtColor(blueChannel, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(7,2))

plt.subplot(131)
plt.title('Red channel')
plt.imshow(redChannel,cmap="gray")
plt.axis("off")

plt.subplot(132)
plt.title("Green Channel")
plt.imshow(greenChannel,cmap="gray")
plt.axis("off")

plt.subplot(133)
plt.title('Blue channel')
plt.imshow(blueChannel,cmap="gray")
plt.axis("off")

plt.savefig('ChannelRepresentationGray.png')
```
