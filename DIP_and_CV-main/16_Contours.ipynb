#What are Contours?

Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

```python
import numpy as np
import cv2 as cv

im = cv.imread('morph.png')
assert im is not None, "file could not be read, check with os.path.exists()"
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

1. To draw all the contours in an image:

```python
cv.drawContours(im, contours, -1, (0,255,0), 3)
```

2. To draw an individual contour, say 4th contour:

```python
cv.drawContours(im, contours, 3, (0,255,0), 3)
```

3. Common Method

```python
cnt = contours[4]
cv.drawContours(im, [cnt], 0, (0,255,0), 3)
```

# Contour Features

###**1. Moments**
used to calculate features like center of mass of the object, area of the object etc.

```python
import numpy as np
import cv2 as cv

img = cv.imread('morph.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
ret,thresh = cv.threshold(img,127,255,0)
contours,hierarchy = cv.findContours(thresh, 1, 2)

cnt = contours[0]
M = cv.moments(cnt)
#print( M ) --> prints the entire dictionary
for i in list(M.items()):
  print(i)
```

The centroid (geometric center) of a shape can be calculated using image moments. If you have the spatial moments of an image or a contour, the centroid coordinates \((cx, cy)\) can be computed as:


$cx = \frac{M_{10}}{M_{00}}, \quad cy = \frac{M_{01}}{M_{00}}$

1. $M_{00}$ (zeroth-order moment): Represents the total area (sum of pixel intensities for binary images).
2. $M_{10}$  and $M_{01}$ (first-order moments): Represent the weighted sums of the x- and y-coordinates, respectively.




```python

cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
print(f"Centroid: ({cx}, {cy})")
```

###**2. Contour Area**

```python
area = cv.contourArea(cnt)
print(f"Area: {area}")
```

###**3. Contour Perimeter**

```python
perimeter = cv.arcLength(cnt,True)
print(f"Perimeter: {perimeter}")
```

###**4. Contour Approximation**

It approximates a contour shape to another shape with less number of vertices depending upon the precision we specify.

Suppose you are trying to find a square in an image, but due to some problems in the image, you didn't get a perfect square, but a "bad shape". Now you can use this function to approximate the shape.

```python
epsilon = 0.1*cv.arcLength(cnt,True)
approx = cv.approxPolyDP(cnt,epsilon,True)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('cap.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Threshold to create a binary image
_, binary = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Get the largest contour (assuming the cap is the main object)
contour = max(contours, key=cv2.contourArea)

# Contour approximation
epsilon = 0.01 * cv2.arcLength(contour, True)  # Adjust epsilon for accuracy
approx = cv2.approxPolyDP(contour, epsilon, True)

# Draw original contour and approximated contour on a copy of the image
output_image = image.copy()
cv2.drawContours(output_image, [contour], -1, (0, 0, 255), 2)  # Original contour in red
cv2.drawContours(output_image, [approx], -1, (0, 255, 0), 2)   # Approximated contour in green

# Display results using matplotlib
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Binary Image")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Contour Approximation")
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

###**5. Convex Hull**

```python
hull = cv.convexHull(cnt)
```

```python
# Draw the original contour and convex hull
output_image = cv2.cvtColor(binary, cv2.COLOR_GRAY2BGR)  # Convert to color for visualization
cv2.drawContours(output_image, [contour], -1, (0, 0, 255), 2)  # Original contour in red
cv2.drawContours(output_image, [hull], -1, (0, 255, 0), 2)     # Convex hull in green

# Display the result
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Binary Image")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Convex Hull")
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

###**7. Bounding Rectangle**
here are two types of bounding rectangles: Straight and rotated

```python
import cv2 as cv
import numpy as np

# Load the image
img = cv.imread('thunder.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Threshold the image to create a binary mask
_, thresh = cv.threshold(img_gray, 127, 255, cv.THRESH_BINARY)

# Find contours
contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

# Loop over all contours
for cnt in contours:
    # Straight bounding rectangle
    x, y, w, h = cv.boundingRect(cnt)
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Green rectangle

    # Rotated bounding rectangle
    rect = cv.minAreaRect(cnt)
    box = cv.boxPoints(rect)
    box = np.int0(box)
    cv.drawContours(img, [box], 0, (0, 0, 255), 2)  # Red rectangle

plt.imshow(img)
plt.show()
```

###**8. Minimum Enclosing Circle**

```python
img = cv.imread('thunder.png')

(x,y),radius = cv.minEnclosingCircle(cnt)
center = (int(x),int(y))
radius = int(radius)
cv.circle(img,center,radius,(0,255,0),2)
```

```python
img = cv.imread('thunder.png')
ellipse = cv.fitEllipse(cnt)
cv.ellipse(img,ellipse,(0,255,0),2)
```

```python
img = cv.imread('thunder.png')

rows,cols = img.shape[:2]
[vx,vy,x,y] = cv.fitLine(cnt, cv.DIST_L2,0,0.01,0.01)
lefty = int((-x*vy/vx) + y)
righty = int(((cols-x)*vy/vx)+y)
cv.line(img,(cols-1,righty),(0,lefty),(0,255,0),2)
```
