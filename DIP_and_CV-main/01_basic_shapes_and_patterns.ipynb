```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a blank image with zeros

def createImage(height, width):
    image = np.zeros((height, width), dtype=np.uint8)
    return image
```

```python
# Function to display image

def showImage(image, title="Image"):
    plt.imshow(image, cmap= "gray")
    plt.title(title)
    plt.axis("Off")
    plt.imsave(title+".png",image, cmap="gray")
```

```python
# Draw a line on the image
# cv.line(img, pt1, pt2, color, thickness) ->	img

image = createImage(200, 200)
lineImage = cv2.line(image, (50, 50), (150, 150), 255, 2)
showImage(lineImage, title="Line")
```

```python
# Draw a rectangle on the image
# cv.rectangle(	img, pt1, pt2, color, thickness) ->	img

image = createImage(200, 200)
rectImage = cv2.rectangle(image, (50, 50), (150, 150), 255, 2)
showImage(rectImage, title="Rectangle")
```

```python
# Draw a circle on the image
# cv.circle(img, center, radius, color, thickness) ->	img


image = createImage(200, 200)
circleImage = cv2.circle(image, (100, 100), 50, 255, 2)
showImage(circleImage, title="Circle")
```

```python
# Draw an ellipse on the image
# cv.ellipse(img, center, axes, angle, startAngle, endAngle, color, thickness) ->	img

image = createImage(200, 200)
ellipseImage = cv2.ellipse(image, (100, 100), (50, 100), 30, 0, 360, 255, 2)
showImage(ellipseImage, title="Ellipse")
```

```python
# Draw a rectangle on the image
# cv.rectangle(	img, pt1, pt2, color, thickness) ->	img

image = createImage(200, 200)
rectImage = cv2.rectangle(image, (100, 50), (150, 150), 255, 2)
showImage(rectImage, title="Rectangle")
```

```python
# Create a blank image with zeros

def createImage(height, width, channels=3):
    image = np.zeros((height, width, channels), dtype=np.uint8)
    return image
```

```python
# Function to display image

def showImage(image, title="Image"):
    plt.imshow(image, cmap= "gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
    plt.imsave(title+".png",image, cmap="gray")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

showImage(image, title="Blank Image")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

# rows 0 to 255
# columns 0 to 127
image[:, :width//2] = 255
showImage(image, title="Half White")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

image[:height//2, :] = 255
showImage(image, title="Half White")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

for i in range(width):
    image[:, i] = i
showImage(image, title="Gradient")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

for i in range(height):
    for j in range(width):
        if(i<j):
            image[i][j] = 255

showImage(image, title="Diagonal")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

for i in range(height):
    for j in range(width):
        if(width-i<j):
            image[i][j] = 255

showImage(image, title="Diagonal")
```

```python
image = createImage(256, 256)
height, width, _ = image.shape

for i in range(height):
    for j in range(width):
        if(i<j):
            if(i<height-j):
                image[i][j] = 255

showImage(image, title="Quadrant")
```

```python
image = createImage(256, 256, 3)
height, width, channels = image.shape

for i in range(height//2):
    for j in range(0, width//2):
        image[i][j] = (255, 0, 0)

showImage(image, title="Red")
```

```python
image = createImage(256, 256, 3)
height, width, channels = image.shape

for i in range(height//2):
    # RED
    for j in range(0, width//2):
        image[i][j] = (255, 0, 0)
    # GREEN
    for j in range(width//2, width):
        image[i][j] = (0, 255, 0)

for i in range(height//2, height):
    # BLUE
    for j in range(0, width//2):
        image[i][j] = (0, 0, 255)

    # WHITE
    for j in range(width//2, width):
        image[i][j] = (255, 255, 255)

showImage(image, title="RGBW")
```

```python
image = createImage(256, 256, 3)
height, width, channels = image.shape

for i in range(height//2):
    # Cyan
    for j in range(0, width//2):
        image[i][j] = (0, 255, 255)
    # Magenta
    for j in range(width//2, width):
        image[i][j] = (255, 0, 255)

for i in range(height//2, height):
    # Yellow
    for j in range(0, width//2):
        image[i][j] = (255, 255, 0)
    # Black
    for j in range(width//2, width):
        image[i][j] = (0, 0, 0)

showImage(image, title="CMYK")
```
